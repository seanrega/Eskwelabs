````python
# TYPE ANSWER HERE
np.random.seed(1337)
foo = np.random.randint(1, 100, size = (10, 3))
foo
````

## Data Manipulation

```python
# 4. Convert `foo` into a Pandas dataframe named `bar`.
#    Set column names to 'spam', 'ham', 'eggs'.
```

````python
# TYPE ANSWER HERE
bar = pd.DataFrame(foo, columns = ['spam', 'ham', 'eggs'])
bar
````

```python
# 5. Select column 'spam' from `bar` and convert it into a Pandas series 
#    named `baz`.
```

````python
# TYPE ANSWER HERE
baz = pd.Series(bar['spam'])
baz
````

```python
# 6. Select column 'ham' and 'eggs' from `bar` and convert it into 
#    Pandas dataframe named `foobar`.
```

````python
# TYPE ANSWER HERE
foobar = bar[['ham', 'eggs']]
foobar
````

## Visualization

```python
# 7. Plot column `eggs` from `foobar` as a bar graph. Add figure title and labels.
```
![](aa_0503_bar.png)

````python
# TYPE ANSWER HERE
foobar['eggs'].plot(kind='bar', rot=0)
plt.title('eggs bar graph')
plt.xlabel('eggs')
plt.ylabel('count')
;
````

```python
# 8. Plot column 'spam' on x-axis and column 'eggs' on y-axis from `bar` as a
#    scatter plot. Add figure title and labels.
```
![](aa_0503_scatter.png)

````python
# TYPE ANSWER HERE
bar[['spam', 'eggs']].plot(kind='scatter', x='spam', y='eggs')
plt.title('Spam and Eggs')
plt.xlabel('spam')
plt.ylabel('eggs')
;
````

```python
# 9. Plot columns 'ham' and 'eggs' from `bar` as box plots.
#    Add figure title and labels.
```
![](aa_0503_box.png)

````python
# TYPE ANSWER HERE
bar[['ham', 'eggs']].plot(kind = 'box')
plt.title('Box ham and eggs')
plt.xlabel('bar')
plt.ylabel('values')
;
````
